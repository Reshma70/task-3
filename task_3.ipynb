```python
import pandas as pd
from google.colab import files
import zipfile
import io
import matplotlib.pyplot as plt

uploaded=files.upload()

for filename in uploaded.keys():
  if filename.endswith('.zip'):
    with zipfile.ZipFile(io.BytesIO(uploaded[filename]), 'r') as zip_ref:
      zip_ref.extractall('.')
      for file_info in zip_ref.infolist():
        if file_info.filename.endswith('.csv'):
          df = pd.read_csv(file_info.filename)
          break
      import pandas as pd


df = pd.read_csv("AB_NYC_2019.csv")
print(f" Initial rows: {df.shape[0]}")


df = df.dropna(subset=['name', 'host_name'])
print(f" Removed rows with missing 'name' or 'host_name'. Remaining rows: {df.shape[0]}")


df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
print(f" Converted 'last_review' to datetime format.")


missing_before = df['reviews_per_month'].isna().sum()
df['reviews_per_month'] = df.apply(
    lambda row: 0 if pd.isna(row['reviews_per_month']) and row['number_of_reviews'] == 0 else row['reviews_per_month'],
    axis=1
)
missing_after = df['reviews_per_month'].isna().sum()
print(f" Imputed 'reviews_per_month': {missing_before - missing_after} values filled with 0.")


df['room_type'] = df['room_type'].str.title().str.strip()
df['neighbourhood_group'] = df['neighbourhood_group'].str.title().str.strip()
print(f" Standardized text columns: 'room_type', 'neighbourhood_group'.")

before_dedup = df.shape[0]
df = df.drop_duplicates()
after_dedup = df.shape[0]
print(f" Removed {before_dedup - after_dedup} duplicate rows.")

df['host_id'] = df['host_id'].astype(int)
df['price'] = df['price'].astype(int)
df['minimum_nights'] = df['minimum_nights'].astype(int)
df['number_of_reviews'] = df['number_of_reviews'].astype(int)
df['calculated_host_listings_count'] = df['calculated_host_listings_count'].astype(int)
df['availability_365'] = df['availability_365'].astype(int)
print(f" Enforced correct numeric datatypes for key columns.")

df.to_csv("AB_NYC_2019_data_integrity_cleaned.csv", index=False)
print(f" Saved final cleaned dataset as 'AB_NYC_2019_data_integrity_cleaned.csv'")

df = df.dropna(subset=['name', 'host_name'])


df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

df['reviews_per_month'] = df.apply(
    lambda row: 0 if pd.isna(row['reviews_per_month']) and row['number_of_reviews'] == 0 else row['reviews_per_month'],
    axis=1
)

missing_report = df.isna().sum()
print(" Missing Value Summary After Handling:\n")
print(missing_report)

duplicates = df.duplicated()
duplicate_count = duplicates.sum()
print(f" Duplicate rows found: {duplicate_count}")

df_cleaned = df.drop_duplicates()
print(f" Dataset shape after duplicate removal: {df_cleaned.shape}")
df_cleaned.to_csv("AB_NYC_2019_no_duplicates.csv", index=False)
print(" Cleaned dataset saved as 'AB_NYC_2019_no_duplicates.csv'")
df['room_type'] = df['room_type'].str.title().str.strip()
df['neighbourhood_group'] = df['neighbourhood_group'].str.title().str.strip()


df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')


for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()


df.to_csv("AB_NYC_2019_standardized.csv", index=False)
print(" Dataset standardized and saved as 'AB_NYC_2019_standardized.csv'")

numeric_cols = ['price', 'minimum_nights', 'number_of_reviews',
                'reviews_per_month', 'calculated_host_listings_count', 'availability_365']


for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    before = df.shape[0]
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    after = df.shape[0]
    print(f" Removed outliers from '{col}': {before - after} rows dropped")


df.to_csv("AB_NYC_2019_no_outliers.csv", index=False)
print(" Final dataset with outliers removed saved as 'AB_NYC_2019_no_outliers.csv'")
```
